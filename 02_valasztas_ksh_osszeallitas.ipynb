```python
## Adatfájl előkészítése az elemzéshez
## Települések választási eredményeinek és különböző mutatóinak összeállítása egy táblázatba

import pandas as pd
fo_tablazat = pd.read_csv("valasztas_eredmenyek.csv")
fo_tablazat
```

```python
## Minden településnél a legtöbb szavazatot kapott párt kigyűjtése
fo_tablazat["legtobb_szavazat"] = fo_tablazat[['DK-JOBBIK-MOMENTUM-MSZP-LMP-PÁRBESZÉD', 'NORMÁLIS PÁRT', 'MKKP', 'MEMO', 'MI HAZÁNK', 'FIDESZ-KDNP', 'Egyéb']].idxmax(axis=1)
```

```python
## Szavazatszámok átalakítása %-os arányra
fo_tablazat = fo_tablazat.assign(szav_ellenzeki_koalicio=lambda x: x['DK-JOBBIK-MOMENTUM-MSZP-LMP-PÁRBESZÉD'] / x['szavazatok_szama'] * 100)
fo_tablazat = fo_tablazat.assign(szav_normalis_part=lambda x: x['NORMÁLIS PÁRT'] / x['szavazatok_szama'] * 100)
fo_tablazat = fo_tablazat.assign(szav_mkkp=lambda x: x['MKKP'] / x['szavazatok_szama'] * 100)
fo_tablazat = fo_tablazat.assign(szav_memo=lambda x: x['MEMO'] / x['szavazatok_szama'] * 100)
fo_tablazat = fo_tablazat.assign(szav_mi_hazank=lambda x: x['MI HAZÁNK'] / x['szavazatok_szama'] * 100)
fo_tablazat = fo_tablazat.assign(szav_fidesz_kdnp=lambda x: x['FIDESZ-KDNP'] / x['szavazatok_szama'] * 100)
fo_tablazat = fo_tablazat.assign(szav_egyeb=lambda x: x['Egyéb'] / x['szavazatok_szama'] * 100)
fo_tablazat = fo_tablazat.assign(reszvetel_szazalek = lambda x: x['szavazatok_szama'] / x['valasztok_szama'] * 100)

fo_tablazat.drop(['DK-JOBBIK-MOMENTUM-MSZP-LMP-PÁRBESZÉD', 'NORMÁLIS PÁRT', 'MKKP', 'MEMO', 'MI HAZÁNK', 'FIDESZ-KDNP', 'Egyéb', 'szavazatok_szama'], inplace=True, axis=1)
fo_tablazat
```

```python
## Település típusok hozzáadása a fő táblázathoz
telepules_tipusok = pd.read_excel("ksh_adatok.xlsx", sheet_name="telepules_tipusok")
telepules_tipusok
```

```python
fo_tablazat = pd.merge(fo_tablazat, telepules_tipusok, how="inner", on="telepulesnev")
fo_tablazat
```

```python
## Megyei és településtípusonkénti szintű mutatók beimportálása, beillesztése a fő táblázatba
erettsegivel_rendelkezok_m = pd.read_excel("ksh_adatok.xlsx", sheet_name="erettsegivel_rendelkezok_m")
fo_tablazat = pd.merge(fo_tablazat, erettsegivel_rendelkezok_m, how="inner", on="megyenev")
fo_tablazat
```

```python
erettsegivel_rendelkezok_t = pd.read_excel("ksh_adatok.xlsx", sheet_name="erettsegivel_rendelkezok_t")
fo_tablazat = pd.merge(fo_tablazat, erettsegivel_rendelkezok_t, how="inner", on="telepulestipus")
fo_tablazat
```

```python
munkanelkuliseg_t = pd.read_excel("ksh_adatok.xlsx", sheet_name="munkanelkuliseg_t")
fo_tablazat = pd.merge(fo_tablazat, munkanelkuliseg_t, how="inner", on="telepulestipus")
fo_tablazat
```

```python
internet_arany_m = pd.read_excel("ksh_adatok.xlsx", sheet_name="internet_arany_m")
fo_tablazat = pd.merge(fo_tablazat, internet_arany_m, how="inner", on="megyenev")
fo_tablazat
```

```python
csalados_haztartasok_m = pd.read_excel("ksh_adatok.xlsx", sheet_name="csalados_haztartasok_m")
fo_tablazat = pd.merge(fo_tablazat, csalados_haztartasok_m, how="inner", on="megyenev")
fo_tablazat
```

```python
csalados_haztartasok_t = pd.read_excel("ksh_adatok.xlsx", sheet_name="csalados_haztartasok_t")
fo_tablazat = pd.merge(fo_tablazat, csalados_haztartasok_t, how="inner", on="telepulestipus")
fo_tablazat
```

```python
## Azoknál a mutatóknál, ahol megyére és településtípusra is van adat, az adott település hovatartozásának megfelelően a kettő átlagát vesszük
## A külön értékeket pedig kitöröljük.
fo_tablazat = fo_tablazat.assign(erettsegivel_rendelkezok=lambda x: (x['erettsegivel_rendelkezok_m']+x['erettsegivel_rendelkezok_t'])/2)
fo_tablazat = fo_tablazat.assign(munkanelkuliseg=lambda x: (x['munkanelkuliseg_rata_m']+x['munkanelkuliseg_rata_t'])/2)
fo_tablazat = fo_tablazat.assign(anyagi_deprivacioban_elok=lambda x: (x['anyagi_deprivacioban_elok_m']+x['anyagi_deprivacioban_elok_t'])/2)
fo_tablazat = fo_tablazat.assign(csalados_haztartasok=lambda x: (x['csalados_haztartasok_m']+x['csalados_haztartasok_t'])/2)
fo_tablazat.drop(['erettsegivel_rendelkezok_m','erettsegivel_rendelkezok_t','munkanelkuliseg_rata_m','munkanelkuliseg_rata_t','anyagi_deprivacioban_elok_m','anyagi_deprivacioban_elok_t',"csalados_haztartasok_m","csalados_haztartasok_t"], axis=1, inplace=True)
fo_tablazat
```

```python
## Elemzésre kész adatok kiexportálása CSV-ként
fo_tablazat.to_csv("valasztas_elemzes.csv", index=False)
```
