```python
## Az egyes településeken mely pártok kapták a legtöbb szavazatot?
legtobb_szavazat = eredmenyek["legtobb_szavazat"].value_counts()
plt.figure(figsize=(16,4))
bar = plt.barh(y=legtobb_szavazat.index, width=legtobb_szavazat.values)
plt.grid(True, axis="x")
plt.bar_label(bar)
plt.tight_layout()
plt.show()
```

```python
## Melyik pártnak kedvezett (vagy fordítva) a magas részvételi arány?
## Mely pártok vontak el leginkább szavazatot egymástól?
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(szazalekos_eredmenyek.corr(), annot=True)
```

```python
## A továbbiakban csak a FIDESZ-KDNP és az ellenézki koalíció eredményeit nézzük
## Sztenderdizáljuk a szavazat %-okat és az elemzéshez használt különböző mutatókat
from sklearn.preprocessing import RobustScaler
elemzes_valtozok = ["szav_ellenzeki_koalicio", "szav_fidesz_kdnp", "valasztok_szama", "haztartas_interneteleressel_m", "atlageletkor_m", "erettsegivel_rendelkezok", "munkanelkuliseg", "anyagi_deprivacioban_elok", "csalados_haztartasok"]
mutato_elemzes = eredmenyek[elemzes_valtozok]
robust_scaler = RobustScaler()
mutato_elemzes[elemzes_valtozok] = robust_scaler.fit_transform(mutato_elemzes[elemzes_valtozok])
mutato_elemzes
```

```python
## Egyes társadalmi mutatók hatása a szavazatok alakulására - Ellenzéki koalició
tarsadalmi_mutatokkal_ellenzek = mutato_elemzes.drop(axis=1, columns="szav_fidesz_kdnp")
plt.figure(figsize=(10,10))
sns.heatmap(tarsadalmi_mutatokkal_ellenzek.corr(), annot=True)
```

```python
## Egyes társadalmi mutatók hatása a szavazatok alakulására - FIDESZ-KDNP
tarsadalmi_mutatokkal_fidesz = mutato_elemzes.drop(axis=1, columns="szav_ellenzeki_koalicio")
plt.figure(figsize=(10,10))
sns.heatmap(tarsadalmi_mutatokkal_fidesz.corr(), annot=True)
```

```python
## Mely telepuléseken szuletett 50% feletti FIDESZ győzelem?
import numpy as np
eredmenyek["fidesz_50_felett"] = np.where(eredmenyek['szav_fidesz_kdnp'] >= 50, "Igen", "Nem")
eredmenyek
```

```python
## Mely településeken alakult 50% feletti FIDESZ győzelem?
legtobb_szavazat = eredmenyek['fidesz_50_felett'].value_counts()
plt.figure(figsize=(16,4))
bar = plt.barh(y=legtobb_szavazat.index, width=legtobb_szavazat.values)
plt.grid(True, axis="x")
plt.bar_label(bar)
plt.tight_layout()
plt.show()
```

```python
## Döntsük el döntési fával, hogy mely településeken várható 50% feletti FIDESZ győzelem

from sklearn.model_selection import train_test_split

fuggo_valtozo = eredmenyek["fidesz_50_felett"]
fuggetlen_valtozok = eredmenyek[tarsadalmi_mutatokkal_fidesz.drop(axis=1, columns="szav_fidesz_kdnp").columns]

## Adatok szétválasztása tanuló és teszt adatokra
x_train, x_test, y_train, y_test = train_test_split(fuggetlen_valtozok, fuggo_valtozo, test_size = 0.2, random_state = 42)
```

```python
## "Egyszerű" döntési fa tanítása, "legjobb mélység megkeresése" (optimalizálás)
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
pontossagok = {}
for x in range(1, 20):
    melyseg = x
    dt = DecisionTreeClassifier(criterion='gini', max_depth=melyseg, random_state = 42)
    dt.fit(x_train, y_train)
    y_pred = dt.predict(x_test)
    pontossag = accuracy_score(y_test,y_pred)
    pontossagok[melyseg] = pontossag

plt.title("Döntési fa pontosságok a döntési fa mélysége szerint")    
plt.bar(pontossagok.keys(), pontossagok.values())
legnagyobb_pontossag_melyseg = max(pontossagok, key=pontossagok.get)
legnagyobb_pontossag_dt = pontossagok[legnagyobb_pontossag_melyseg]
print("Legnagyobb pontossághoz szükséges mélység: " + str(legnagyobb_pontossag_melyseg))
print("Legnagyobb pontosság: " + str(legnagyobb_pontossag_dt))
```

```python
## Ábrázoljuk az "optimális" döntési fát
from sklearn import tree
dt = DecisionTreeClassifier(criterion='gini', max_depth=legnagyobb_pontossag_melyseg, random_state = 42)
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
plt.figure(figsize=(30,15))
tree.plot_tree(dt, filled=True, rounded=True, feature_names=x_train.columns, fontsize=12)
plt.show()
```

```python
## Segédmetódus a valódi és predikált értékek ábrázolására
def abrazolas(title, y_test, y_pred, pontossag):
    plt.figure(figsize=(10,5))
    plt.scatter(range(len(y_pred)), y_pred, color="red", label="Predikált értékek")
    plt.scatter(range(len(y_test)), y_test, color="green", label="Valódi értékek")
    plt.suptitle(title, fontsize=15)
    plt.title("Pontosság: " + str(pontossag*100) + "%", fontsize=10)
    plt.legend()
    plt.show()
```

```python
## Logisztikus regresszió előrejelzéseinek ábrázolása
abrazolas("Döntési fa", y_test, y_pred, accuracy_score(y_test,y_pred))
```
